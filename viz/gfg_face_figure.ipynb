```python
import os
os.environ['DISPLAY'] = ':0.0'

from GFG import Ctx
from GFG.model import Nf, Adata
```

```python
adata = Adata.load('quick_FACS_blendshapes_v2dense')
base_nf = Nf.from_default()
```

```python
ctx = Ctx(hidden=True)
ctx.backgroundcolor = [1., 1., 1., 0.]
base_nf.attach(ctx)
adata.attach(ctx)
ctx.set_lights('flat')
ctx.devmapcm('Reds')
```

## Figure 4A
First, 'disgust' with AU9 and AU25

```python
adata.bshapes['AU9'] = 1.
adata.bshapes['AU25'] = 1.
```

```python
ctx.devmapwt = 0.
img = ctx.render('image')
img.save('./AU9+25_nocolor.png')
img
```

```python
adata.bshapes['AU9'] = 1.
adata.bshapes['AU25'] = 0.

ctx.devmapwt = 1.
ctx.devmapmx = 12.
ctx.devmapmn = 0.
img = ctx.render('image')
img.save('./AU9_red.png')
img
```

```python
adata.bshapes['AU9'] = 1.
adata.bshapes['AU25'] = 0.
ctx.devmapwt = 0.
img = ctx.render('image')
img.save('./AU9_nocolor.png')
img
```

```python
adata.bshapes['AU9'] = 0.
adata.bshapes['AU25'] = 1.
ctx.devmapcm('Blues')
ctx.devmapwt = 1.
ctx.devmapmx = 4.
ctx.devmapmn = 0.
img = ctx.render('image')
img.save('./AU25_blue.png')
img
```

```python
adata.bshapes['AU9'] = 0.
adata.bshapes['AU25'] = 1.
ctx.devmapwt = 0.
img = ctx.render('image')
img.save('./AU25_nocolor.png')
img
```
