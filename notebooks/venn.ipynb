```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from matplotlib.patches import Wedge, Circle
from matplotlib.collections import PatchCollection
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
v = venn2(
    {'10': 4, '11': 2, '01': 4},
    set_labels=['', ''],
    alpha=1,
    normalize_to=10,
    ax=ax,
    subset_label_formatter=lambda x: ''
)
#v.get_patch_by_id('10').set_color('white')
v.get_patch_by_id('10').set_alpha(0)
v.get_patch_by_id('11').set_color('#ff9900')
v.get_patch_by_id('01').set_color('#4a86e8')

w = Wedge(center=(-0.755, 0), r=1.385, theta1=179.9, theta2=0, alpha=1, fc='#cc4125', zorder=-1)
ax.add_patch(w)
w = Wedge(center=(-0.755, 0), r=1.385, theta1=0, theta2=180.1, alpha=1, fc='#6aa84f', zorder=-1)
ax.add_patch(w)
c = Circle((-0.755, 0), radius=1.385, alpha=1, ls='--', lw=3, ec='k', fc='none')
ax.add_patch(c)
```
